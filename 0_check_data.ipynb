# Importing the data
Just a quick check to see whether we can import all the data

```python
import tmd
from tmd.view import plot
from tmd.view import view

import matplotlib.pyplot as plt
import os
import functions as fcts
```

```python
data_folder = "Data/ExamplesL23PC"
folders_to_treat = ['Human','Mouse']
```

```python
grouped_neurons = fcts.check_and_load_neurons(data_folder, folders_to_treat)
```

```python
neurons_human = grouped_neurons["Human"]
neurons_mouse = grouped_neurons["Mouse"]
```

# Checking the average diagram
Depending on the part of the neurons: all neurons, apical dendrite, basale dendrite, axon.

## Whole neurons

```python
ph_diagrams_human_whole = fcts.compute_diagrams(
    neurons_human, 
    group_name="Human", 
    neuron_part="dendrites",
    distance_type="path_distances"
)

ph_diagrams_mouse_whole = fcts.compute_diagrams(
    neurons_mouse, 
    group_name="Mouse", 
    neuron_part="dendrites",
    distance_type="path_distances"
)
```

```python
human_whole = plot.persistence_image_average(ph_diagrams_human_whole)
```

```python
mouse_whole = plot.persistence_image_average(ph_diagrams_mouse_whole)
```
